```python
import random

daftar_pola_soal = [
    {"template": "{nama} punya {a} apel, lalu diberi {b} apel lagi. Berapa jumlah apel {nama} sekarang?", "operasi": "+"},
    {"template": "{nama} punya {a} permen, lalu memakan {b} permen. Berapa sisa permen {nama}?", "operasi": "-"},
    {"template": "{nama} membeli {a} bungkus kerupuk, setiap bungkus berisi {b} kerupuk. Berapa total kerupuk yang dibeli {nama}?", "operasi": "*"},
    {"template": "{nama} punya {a} kelereng, lalu menemukan {b} kelereng lagi di halaman. Berapa kelereng {nama} sekarang?", "operasi": "+"},
    {"template": "Ibu {nama} membagikan {a} kue ke {b} orang secara rata. Berapa kue yang diterima tiap orang?", "operasi": "/"},
]

nama_pemain_list = ["Budi", "Siti", "Andi", "Dewi", "Rian", "Ayu"]

komentar_benar = ["🎉 Keren banget!", "👏 Betul sekali!", "🌟 Kamu jago matematika!", "💯 Mantap!"]
komentar_salah = ["😅 Hampir, coba lagi ya!", "🤔 Belum tepat nih.", "📚 Yuk dihitung lagi pelan-pelan."]


def buat_soal():
    pola = random.choice(daftar_pola_soal)
    nama = random.choice(nama_pemain_list)

    if pola["operasi"] == "/":
        # supaya hasil bagi selalu bilangan bulat yang rapi
        b = random.randint(2, 5)
        jawaban = random.randint(2, 10)
        a = b * jawaban
    else:
        a = random.randint(2, 10)
        b = random.randint(1, 5)
        if pola["operasi"] == "+":
            jawaban = a + b
        elif pola["operasi"] == "-":
            a = max(a, b + 1)  # supaya hasil pengurangan tidak negatif
            jawaban = a - b
        elif pola["operasi"] == "*":
            jawaban = a * b

    teks_soal = pola["template"].format(nama=nama, a=a, b=b)
    return teks_soal, jawaban


def main_kalkulator_cerita():
    jumlah_soal = 5
    skor = 0

    print("=" * 50)
    print("🧮 SELAMAT DATANG DI KALKULATOR CERITA MATEMATIKA!")
    print("=" * 50)
    print(f"Ada {jumlah_soal} soal cerita menantimu. Semangat! 💪\n")

    for nomor in range(1, jumlah_soal + 1):
        soal, jawaban_benar = buat_soal()
        print(f"Soal {nomor}: {soal}")

        try:
            jawaban_pemain = int(input("Jawabanmu: "))
        except ValueError:
            print("⚠️  Itu bukan angka, dianggap salah ya.\n")
            continue

        if jawaban_pemain == jawaban_benar:
            print(random.choice(komentar_benar), "\n")
            skor += 1
        else:
            print(random.choice(komentar_salah), f"Jawaban yang benar: {jawaban_benar}\n")

    print("=" * 50)
    print(f"🏁 Selesai! Skormu: {skor}/{jumlah_soal}")

    if skor == jumlah_soal:
        print("🏆 Sempurna! Kamu jagoan matematika!")
    elif skor >= jumlah_soal / 2:
        print("👍 Bagus! Terus berlatih ya!")
    else:
        print("💪 Jangan menyerah, coba lagi!")


main_kalkulator_cerita()
```
