```python
daftar_barang = {
    "beras": 12000,
    "gula": 14000,
    "telur": 2000,
    "mie instan": 3000,
    "minyak goreng": 16000,
    "sabun": 4000,
}


def format_rupiah(angka):
    return "Rp" + f"{angka:,}".replace(",", ".")


def tampilkan_daftar_barang():
    print("=" * 35)
    print("       DAFTAR BARANG TOKO")
    print("=" * 35)
    for nama_barang, harga in daftar_barang.items():
        print(f"{nama_barang.title():<18} {format_rupiah(harga)}")
    print("=" * 35)


def hitung_total(keranjang):
    total = 0
    for barang in keranjang:
        total += barang["jumlah"] * barang["harga_satuan"]
    return total


def cetak_struk(keranjang, total, uang_dibayar, kembalian):
    print("\n" + "=" * 35)
    print("           STRUK BELANJA")
    print("=" * 35)
    for barang in keranjang:
        subtotal = barang["jumlah"] * barang["harga_satuan"]
        print(f"{barang['nama'].title():<15} x{barang['jumlah']:<3} {format_rupiah(subtotal)}")
    print("-" * 35)
    print(f"{'Total':<19} {format_rupiah(total)}")
    print(f"{'Uang Dibayar':<19} {format_rupiah(uang_dibayar)}")
    print(f"{'Kembalian':<19} {format_rupiah(kembalian)}")
    print("=" * 35)
    print("     Terima kasih telah belanja! 🙏")
    print("=" * 35)


def main_toko_kelontong():
    print("🏪 SELAMAT DATANG DI TOKO KELONTONG PYTHON!\n")
    tampilkan_daftar_barang()

    keranjang = []

    while True:
        nama_barang = input("\nMasukkan nama barang (atau ketik 'selesai' untuk membayar): ").strip().lower()

        if nama_barang == "selesai":
            break

        if nama_barang not in daftar_barang:
            print(f"⚠️  Maaf, barang '{nama_barang}' tidak tersedia di toko kami.")
            continue

        try:
            jumlah = int(input(f"Berapa jumlah '{nama_barang}' yang ingin dibeli? "))
        except ValueError:
            print("⚠️  Jumlah harus berupa angka!")
            continue

        if jumlah <= 0:
            print("⚠️  Jumlah harus lebih dari 0!")
            continue

        keranjang.append({
            "nama": nama_barang,
            "jumlah": jumlah,
            "harga_satuan": daftar_barang[nama_barang]
        })
        subtotal = jumlah * daftar_barang[nama_barang]
        print(f"✅ {jumlah} {nama_barang} ditambahkan ke keranjang ({format_rupiah(subtotal)})")

    if not keranjang:
        print("\nKeranjang masih kosong. Sampai jumpa lagi! 👋")
        return

    total = hitung_total(keranjang)
    print(f"\nTotal belanjaanmu: {format_rupiah(total)}")

    while True:
        try:
            uang_dibayar = int(input("Masukkan jumlah uang yang dibayarkan: "))
        except ValueError:
            print("⚠️  Masukkan angka yang valid!")
            continue

        if uang_dibayar < total:
            kekurangan = total - uang_dibayar
            print(f"⚠️  Uangmu kurang {format_rupiah(kekurangan)}. Coba masukkan jumlah lain.")
            continue
        break

    kembalian = uang_dibayar - total
    cetak_struk(keranjang, total, uang_dibayar, kembalian)


main_toko_kelontong()
```
