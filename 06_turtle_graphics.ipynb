```python
import turtle
import random

layar = turtle.Screen()
layar.title("Kebun Bunga Python")
layar.bgcolor("skyblue")

kura = turtle.Turtle()
kura.speed(0)   # 0 artinya secepat mungkin


def gambar_bunga(kura, ukuran, warna):
    """Menggambar satu bunga sederhana dari kelopak-kelopak berbentuk lingkaran kecil."""
    kura.pencolor(warna)
    kura.fillcolor(warna)

    for i in range(8):   # 8 kelopak bunga
        kura.begin_fill()
        kura.circle(ukuran, steps=6)
        kura.end_fill()
        kura.left(45)


warna_bunga = ["pink", "red", "purple", "orange", "yellow"]

for i in range(6):
    kura.penup()
    x = random.randint(-250, 250)
    y = random.randint(-150, 150)
    kura.goto(x, y)
    kura.pendown()

    ukuran = random.randint(15, 30)
    warna = random.choice(warna_bunga)
    gambar_bunga(kura, ukuran, warna)

kura.hideturtle()   # sembunyikan bentuk kura-kura setelah selesai menggambar
print("Kebun bunga selesai digambar! 🌸")
```
