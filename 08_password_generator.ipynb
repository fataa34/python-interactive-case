```python
import random
import string


def buat_password(panjang=8):
    semua_karakter = string.ascii_lowercase + string.ascii_uppercase + string.digits + string.punctuation
    return "".join(random.choice(semua_karakter) for i in range(panjang))


def cek_kekuatan_password(password):
    skor = 0
    if len(password) >= 8:
        skor += 1
    if len(password) >= 12:
        skor += 1
    if any(k.islower() for k in password):
        skor += 1
    if any(k.isupper() for k in password):
        skor += 1
    if any(k.isdigit() for k in password):
        skor += 1
    if any(k in string.punctuation for k in password):
        skor += 1

    if skor <= 2:
        label = "🔴 Lemah"
    elif skor <= 4:
        label = "🟡 Sedang"
    else:
        label = "🟢 Kuat"

    return skor, label


def main_kode_rahasia():
    print("=" * 45)
    print("   🔐 PEMBUAT & PENGECEK KODE RAHASIA 🔐")
    print("=" * 45)
    print("1. Buat password baru secara otomatis")
    print("2. Cek kekuatan password milikku sendiri")
    print("3. Keluar")

    pilihan = input("\nPilih menu (1/2/3): ").strip()

    if pilihan == "1":
        try:
            panjang = int(input("Mau berapa panjang passwordnya? (misal 12): "))
        except ValueError:
            panjang = 12
            print("Input tidak valid, memakai panjang default 12.")

        password_baru = buat_password(panjang)
        skor, label = cek_kekuatan_password(password_baru)
        print(f"\n✨ Password barumu: {password_baru}")
        print(f"Kekuatan: {label} (skor {skor}/6)")

    elif pilihan == "2":
        password_input = input("Masukkan password yang ingin dicek: ")
        skor, label = cek_kekuatan_password(password_input)
        print(f"\nKekuatan password '{password_input}': {label} (skor {skor}/6)")

        if skor <= 2:
            print("💡 Tip: tambahkan huruf besar, angka, dan simbol supaya lebih kuat!")
        elif skor <= 4:
            print("💡 Tip: coba buat lebih panjang lagi (minimal 12 karakter) untuk lebih aman!")
        else:
            print("👍 Password ini sudah cukup kuat!")

    elif pilihan == "3":
        print("Sampai jumpa, agen rahasia! 🕵️")

    else:
        print("⚠️  Pilihan tidak dikenali.")


main_kode_rahasia()
```
