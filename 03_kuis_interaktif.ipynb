```python
daftar_soal = [
    {
        "pertanyaan": "Ibu kota Indonesia adalah?",
        "pilihan": {"A": "Bandung", "B": "Jakarta", "C": "Surabaya", "D": "Medan"},
        "jawaban_benar": "B"
    },
    {
        "pertanyaan": "Planet terdekat dari Matahari adalah?",
        "pilihan": {"A": "Bumi", "B": "Venus", "C": "Merkurius", "D": "Mars"},
        "jawaban_benar": "C"
    },
    {
        "pertanyaan": "Hasil dari 7 x 8 adalah?",
        "pilihan": {"A": "54", "B": "56", "C": "58", "D": "64"},
        "jawaban_benar": "B"
    },
    {
        "pertanyaan": "Hewan apa yang dikenal sebagai 'raja hutan'?",
        "pilihan": {"A": "Harimau", "B": "Gajah", "C": "Singa", "D": "Serigala"},
        "jawaban_benar": "C"
    },
    {
        "pertanyaan": "Berapa jumlah hari dalam satu tahun kabisat?",
        "pilihan": {"A": "365", "B": "364", "C": "366", "D": "360"},
        "jawaban_benar": "C"
    },
]


def tampilkan_soal(nomor, soal):
    print(f"\nSoal {nomor}: {soal['pertanyaan']}")
    for kode, teks in soal["pilihan"].items():
        print(f"  {kode}. {teks}")


def beri_gelar(skor, total_soal):
    persentase = skor / total_soal * 100
    if persentase == 100:
        return "🏆 Sang Juara! Nilai sempurna!"
    elif persentase >= 80:
        return "🌟 Hebat sekali, hampir sempurna!"
    elif persentase >= 60:
        return "👍 Bagus, terus berlatih ya!"
    elif persentase >= 40:
        return "📚 Lumayan, ayo belajar lagi sedikit!"
    else:
        return "💪 Jangan menyerah, coba lagi lain kali!"


def main_kuis():
    skor = 0
    total_soal = len(daftar_soal)

    print("=" * 50)
    print("🧠 SELAMAT DATANG DI KUIS INTERAKTIF!")
    print("=" * 50)
    print(f"Ada {total_soal} soal untukmu. Semangat menjawab! 💪")

    for nomor, soal in enumerate(daftar_soal, start=1):
        tampilkan_soal(nomor, soal)
        jawaban_pemain = input("Jawabanmu (A/B/C/D): ").strip().upper()

        if jawaban_pemain == soal["jawaban_benar"]:
            print("✅ Benar! 🎉")
            skor += 1
        else:
            jawaban_benar_teks = soal["pilihan"][soal["jawaban_benar"]]
            print(f"❌ Kurang tepat. Jawaban yang benar: {soal['jawaban_benar']}. {jawaban_benar_teks}")

    print("\n" + "=" * 50)
    print(f"🏁 Kuis selesai! Skor akhirmu: {skor}/{total_soal}")
    print(beri_gelar(skor, total_soal))


main_kuis()
```
