```python
from datetime import date
import random


def hitung_umur(tanggal_lahir, hari_ini):
    umur = hari_ini.year - tanggal_lahir.year
    if (hari_ini.month, hari_ini.day) < (tanggal_lahir.month, tanggal_lahir.day):
        umur -= 1
    return umur


def hitung_hari_menuju_ultah(tanggal_lahir, hari_ini):
    bulan, tanggal = tanggal_lahir.month, tanggal_lahir.day

    # Menangani kasus lahir 29 Februari di tahun bukan kabisat
    try:
        ultah_tahun_ini = date(hari_ini.year, bulan, tanggal)
    except ValueError:
        ultah_tahun_ini = date(hari_ini.year, 3, 1)   # dirayakan 1 Maret kalau bukan tahun kabisat

    if ultah_tahun_ini < hari_ini:
        try:
            ultah_berikutnya = date(hari_ini.year + 1, bulan, tanggal)
        except ValueError:
            ultah_berikutnya = date(hari_ini.year + 1, 3, 1)
    else:
        ultah_berikutnya = ultah_tahun_ini

    selisih = ultah_berikutnya - hari_ini
    return selisih.days, ultah_berikutnya


def pesan_ulang_tahun(hari_lagi):
    pesan_hari_h = [
        "🎉🎂 SELAMAT ULANG TAHUN! Semoga semua harapanmu terkabul!",
        "🥳 Hari ini harimu! Selamat merayakan bertambahnya usia!",
        "🎈 Yeay, hari ini ulang tahunmu! Semoga tahun ini penuh kejutan seru!",
    ]
    pesan_dekat = [
        "😆 Wah, tinggal {} hari lagi! Sudah siap-siap kue belum?",
        "🎁 {} hari lagi menuju hari spesialmu, sudah bikin daftar hadiah?",
        "⏳ Hitung mundur dimulai! {} hari lagi ya!",
    ]
    pesan_masih_jauh = [
        "📅 Masih {} hari lagi nih, sabar ya sambil menabung untuk pesta!",
        "🕰️ {} hari lagi, masih lama tapi pasti seru nanti!",
        "🌱 {} hari menuju ulang tahun, waktunya merencanakan sesuatu yang seru!",
    ]

    if hari_lagi == 0:
        return random.choice(pesan_hari_h)
    elif hari_lagi <= 7:
        return random.choice(pesan_dekat).format(hari_lagi)
    else:
        return random.choice(pesan_masih_jauh).format(hari_lagi)


def main_kalender_ultah():
    print("=" * 45)
    print("   🎂 KALENDER ULANG TAHUN & REMINDER 🎂")
    print("=" * 45)

    try:
        tahun = int(input("Tahun lahirmu (contoh 2014): "))
        bulan = int(input("Bulan lahirmu (1-12): "))
        tanggal = int(input("Tanggal lahirmu (1-31): "))
        tanggal_lahir = date(tahun, bulan, tanggal)
    except ValueError:
        print("⚠️  Tanggal yang dimasukkan tidak valid. Coba jalankan ulang ya!")
        return

    hari_ini = date.today()

    if tanggal_lahir > hari_ini:
        print("⚠️  Tanggal lahir tidak boleh di masa depan!")
        return

    umur = hitung_umur(tanggal_lahir, hari_ini)
    hari_lagi, ultah_berikutnya = hitung_hari_menuju_ultah(tanggal_lahir, hari_ini)

    print(f"\n📆 Hari ini: {hari_ini}")
    print(f"🎈 Umurmu sekarang: {umur} tahun")
    print(f"🎁 Ulang tahun berikutnya: {ultah_berikutnya}")
    print(f"\n{pesan_ulang_tahun(hari_lagi)}")


main_kalender_ultah()
```
