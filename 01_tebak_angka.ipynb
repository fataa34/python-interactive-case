```python
import random

def main_tebak_angka():
    angka_rahasia = random.randint(1, 100)
    maksimal_percobaan = 7
    percobaan_ke = 0

    print("=" * 40)
    print("SELAMAT DATANG DI GAME TEBAK ANGKA!")
    print("=" * 40)
    print(f"Aku sudah memikirkan angka antara 1-100.")
    print(f"Kamu punya {maksimal_percobaan} kali kesempatan. Ayo tebak!\n")

    while percobaan_ke < maksimal_percobaan:
        percobaan_ke += 1
        sisa_kesempatan = maksimal_percobaan - percobaan_ke

        try:
            tebakan = int(input(f"Percobaan {percobaan_ke}/{maksimal_percobaan} - Tebakanmu: "))
        except ValueError:
            print("Itu bukan angka yang valid, coba lagi ya!")
            percobaan_ke -= 1  # percobaan tidak dihitung kalau input salah
            continue

        if tebakan == angka_rahasia:
            print(f"\n SELAMAT! Kamu berhasil menebak angka {angka_rahasia}!")
            print(f"Kamu berhasil dalam {percobaan_ke} kali percobaan. Hebat")
            return
        elif tebakan < angka_rahasia:
            print(f"Terlalu kecil! Sisa kesempatan: {sisa_kesempatan}\n")
        else:
            print(f"Terlalu besar! Sisa kesempatan: {sisa_kesempatan}\n")

    print(f"\n😢 Yah, kesempatanmu habis. Angka rahasianya adalah {angka_rahasia}.")
    print("Coba lagi yuk, jalankan cell ini sekali lagi!")


main_tebak_angka()
```
