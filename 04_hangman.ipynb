```python
import random

bank_kata = {
    "hewan": ["gajah", "jerapah", "kucing", "harimau", "kelinci", "buaya"],
    "buah": ["mangga", "rambutan", "nanas", "semangka", "durian", "apel"],
    "tokoh kartun": ["spongebob", "doraemon", "upin", "nobita", "shinchan"],
}

gambar_nyawa = [
    "\n   ------\n   |    |\n   |\n   |\n   |\n   |\n---------",
    "\n   ------\n   |    |\n   |    O\n   |\n   |\n   |\n---------",
    "\n   ------\n   |    |\n   |    O\n   |    |\n   |\n   |\n---------",
    "\n   ------\n   |    |\n   |    O\n   |   /|\n   |\n   |\n---------",
    "\n   ------\n   |    |\n   |    O\n   |   /|\\\n   |\n   |\n---------",
    "\n   ------\n   |    |\n   |    O\n   |   /|\\\n   |   /\n   |\n---------",
    "\n   ------\n   |    |\n   |    O\n   |   /|\\\n   |   / \\\n   |\n---------",
]


def tampilkan_kata(kata_rahasia, huruf_ditebak):
    tampilan = [huruf if huruf in huruf_ditebak else "_" for huruf in kata_rahasia]
    return " ".join(tampilan)


def main_hangman():
    kategori = random.choice(list(bank_kata.keys()))
    kata_rahasia = random.choice(bank_kata[kategori])

    huruf_ditebak = set()
    nyawa_maksimal = 6
    kesalahan = 0

    print("=" * 50)
    print("🔤 SELAMAT DATANG DI GAME HANGMAN!")
    print("=" * 50)
    print(f"Kategori kata: {kategori.upper()}")
    print(f"Kamu punya {nyawa_maksimal} kali kesempatan salah.\n")

    while kesalahan < nyawa_maksimal:
        print(gambar_nyawa[kesalahan])
        print("\nKata:", tampilkan_kata(kata_rahasia, huruf_ditebak))
        print("Huruf yang sudah ditebak:", ", ".join(sorted(huruf_ditebak)) if huruf_ditebak else "(belum ada)")

        tebakan = input("\nTebak satu huruf: ").strip().lower()

        if len(tebakan) != 1 or not tebakan.isalpha():
            print("⚠️  Masukkan satu huruf saja ya!\n")
            continue

        if tebakan in huruf_ditebak:
            print("🔁 Huruf itu sudah kamu tebak sebelumnya, coba huruf lain!\n")
            continue

        huruf_ditebak.add(tebakan)

        if tebakan in kata_rahasia:
            print(f"✅ Benar! Huruf '{tebakan}' ada di dalam kata.\n")
        else:
            kesalahan += 1
            print(f"❌ Yah, huruf '{tebakan}' tidak ada. Sisa nyawa: {nyawa_maksimal - kesalahan}\n")

        # Cek kemenangan
        if set(kata_rahasia).issubset(huruf_ditebak):
            print("🎉🎉 SELAMAT! Kamu berhasil menebak kata:", kata_rahasia.upper())
            return

    # Kalau keluar loop karena nyawa habis
    print(gambar_nyawa[nyawa_maksimal])
    print(f"\n😢 Yah, nyawamu habis. Kata rahasianya adalah: {kata_rahasia.upper()}")
    print("Coba lagi yuk, jalankan cell ini sekali lagi!")


main_hangman()
```
